#**Calculadora Simples**

```python
try:
    # Solicita ao usuário o primeiro número e o converte para float para permitir decimais.
    num1 = float(input("Digite o primeiro número: "))
    # Solicita ao usuário o segundo número e o converte para float.
    num2 = float(input("Digite o segundo número: "))
    # Solicita a operação desejada e a converte para minúsculas para padronização.
    operacao = input("Escolha a operação (somar ou subtrair): ").lower()

    # Verifica qual operação foi escolhida pelo usuário.
    if operacao == "somar":
        # Executa a operação de soma e armazena o resultado.
        resultado = num1 + num2
        # Imprime o resultado da soma formatado.
        print(f"A soma é: {resultado}")
    elif operacao == "subtrair":
        # Executa a operação de subtração e armazena o resultado.
        resultado = num1 - num2
        # Imprime o resultado da subtração formatado.
        print(f"A subtração é: {resultado}")
    else:
        # Caso a operação seja inválida, informa o usuário.
        print("Operação inválida. Por favor, escolha 'somar' ou 'subtrair'.")
# Captura exceções caso a entrada do usuário não possa ser convertida para float (não é um número).
except ValueError:
    # Informa ao usuário que a entrada foi inválida.
    print("Entrada inválida. Por favor, digite apenas números.")
```
