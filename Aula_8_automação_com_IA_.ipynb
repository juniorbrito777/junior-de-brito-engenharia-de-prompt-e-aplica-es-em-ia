```python
# Este script Python simples verifica a temperatura de um sistema.
# Se a temperatura for superior a 30 graus Celsius, ele exibirá um alerta de superaquecimento.

def verificar_temperatura(temperatura):
    """
    Verifica a temperatura fornecida e imprime uma mensagem apropriada.
    Se a temperatura for superior a 30°C, um alerta de superaquecimento é exibido.

    Args:
        temperatura (int ou float): O valor da temperatura a ser verificado.
    """
    # Verifica se a temperatura é maior que 30 graus Celsius
    if temperatura > 30:
        # Exibe um alerta de superaquecimento se a condição for verdadeira
        print("ALERTA: Temperatura de " + str(temperatura) + "°C detectada. Superaquecimento!")
    else:
        # Caso contrário, indica que a temperatura está normal
        print("Temperatura atual: " + str(temperatura) + "°C. Tudo normal.")

# --- Exemplos de uso da função ---

# Exemplo 1: Temperatura normal
# Defina uma temperatura que não acione o alerta.
# Você pode substituir este valor pela leitura de um sensor real, se disponível.
temperatura_atual = 56
verificar_temperatura(temperatura_atual)

# Exemplo 2: Temperatura alta (aciona o alerta)
# Defina uma temperatura que seja maior que 30 para testar o alerta.
temperatura_alta = 35
verificar_temperatura(temperatura_alta)
```
